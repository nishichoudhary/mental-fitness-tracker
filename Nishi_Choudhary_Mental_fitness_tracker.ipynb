Importing Libraries

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df1 =  pd.read_csv('/content/drive/MyDrive/Mental fitness tracker/prevalence-by-mental-and-substance-use-disorder _AI.csv')
df2 = pd.read_csv('/content/drive/MyDrive/Mental fitness tracker/mental-and-substance-use-as-share-of-disease -AI.csv')

# merging 2 datasets
data = pd.merge(df1, df2)
data.head(10)
```

```python
# Transform non-numeric lables to numeric lables

from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in data.columns:
  if data[i].dtype=='object':
    data[i]=l.fit_transform(data[i])
```

Split Data

In this step we are going to split the data in 2 parts(training and testing), so that we can train our model on training dataset and test its accuracy on unseen test data.

```python
x=data.drop('Mental_fitness',axis=1)
y= data['Mental_fitness']
from sklearn.model_selection import train_test_split          # used to split original data into training and test data
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=.20, random_state=2)

print('xtrain: ', xtrain.shape)
print('xtest: ', xtest.shape)
print('\n ytrain: ', ytrain.shape)
print('ytest: ', ytest.shape)
```

1) Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
lr= LinearRegression()
lr.fit(xtrain,ytrain)

ytrain_pred = lr.predict(xtrain)
mse = mean_squared_error(ytrain, ytrain_pred)
rmse = (np.sqrt(mean_squared_error(ytrain, ytrain_pred)))
r2 = r2_score(ytrain, ytrain_pred)
print("-----Linear Regression Model for Training Set------")
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
ytest_pred = lr.predict(xtest)
mse = mean_squared_error(ytest, ytest_pred)
rmse = (np.sqrt(mean_squared_error(ytest, ytest_pred)))
r2 = r2_score(ytest, ytest_pred)
print("\n-----Linear Regression Model for Testing Set------")
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```

2)  Random Forest Regressor Model

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(xtrain, ytrain)

ytrain_pred = rf.predict(xtrain)
mse = mean_squared_error(ytrain, ytrain_pred)
rmse = (np.sqrt(mean_squared_error(ytrain, ytrain_pred)))
r2 = r2_score(ytrain, ytrain_pred)
print('---------Random Forest Regressor Model for Training set------')
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))

ytest_pred = rf.predict(xtest)
mse = mean_squared_error(ytest, ytest_pred)
rmse = (np.sqrt(mean_squared_error(ytest, ytest_pred)))
r2 = r2_score(ytest, ytest_pred)
print("\n-----Random Forest Regressor Model for Testing Set------")
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```

3)  SVM Regression

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
svm = SVR()
svm.fit(xtrain, ytrain)

ytrain_pred = svm.predict(xtrain)
mse_train = mean_squared_error(ytrain, ytrain_pred)
rmse_train = np.sqrt(mean_squared_error(ytrain, ytrain_pred))
r2_train = r2_score(ytrain, ytrain_pred)
print('---------Support Vector Machine Regression Model for Training Set------')
print('MSE is {}'.format(mse_train))
print('RMSE is {}'.format(rmse_train))
print('R2 score is {}'.format(r2_train))
ytest_pred = svm.predict(xtest)
mse_test = mean_squared_error(ytest, ytest_pred)
rmse_test = np.sqrt(mean_squared_error(ytest, ytest_pred))
r2_test = r2_score(ytest, ytest_pred)
print("\n-----Support Vector Machine Regression Model for Testing Set------")
print('MSE is {}'.format(mse_test))
print('RMSE is {}'.format(rmse_test))
print('R2 score is {}'.format(r2_test))
```

4)  Decision Tree Regressor Model

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(xtrain, ytrain)

ytrain_pred = dt.predict(xtrain)
mse_train = mean_squared_error(ytrain, ytrain_pred)
rmse_train = np.sqrt(mean_squared_error(ytrain, ytrain_pred))
r2_train = r2_score(ytrain, ytrain_pred)
print('---------Decision Tree Regressor Model for Training Set------')
print('MSE is {}'.format(mse_train))
print('RMSE is {}'.format(rmse_train))
print('R2 score is {}'.format(r2_train))

ytest_pred = dt.predict(xtest)
mse_test = mean_squared_error(ytest, ytest_pred)
rmse_test = np.sqrt(mean_squared_error(ytest, ytest_pred))
r2_test = r2_score(ytest, ytest_pred)
print("\n-----Decision Tree Regressor Model for Testing Set------")
print('MSE is {}'.format(mse_test))
print('RMSE is {}'.format(rmse_test))
print('R2 score is {}'.format(r2_test))
```

Conclusion:

Random Forest Regression works well on both train and test sets with r2 score of 0.99.
As well as Decision Tree Regression also works well on both train and test set with r2 score of 0.98.
